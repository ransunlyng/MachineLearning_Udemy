```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
#Functional Method
```

```python
plt.plot(x,y,'r-')
plt.show()
```

```python
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')

```

```python
#Object oriented method
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x,y)
axes1.set_title('Larger plot')
axes2.plot(y,x)
axes2.set_title('Smaller plot')
axes2.set_xlabel('X Label')
```

```python
fig, axes = plt.subplots(nrows = 3, ncols = 3)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2) #plt.subplot is doing plt.figure(), then fig.add_axes([]) for you, based on how many rows and columns you want

for current_ax in axes:
    current_ax.plot(x, y)
```

```python
axes # axes is a list that you can iterate through. We can also use index on it. since it is a list
```

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')
plt.tight_layout()
```

```python
fig1 = plt.figure(figsize=(8,2), dpi=100)

ax = fig1.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
x = np.linspace(1, 10, 100)
y = np.log(x)

fig2, axes = plt.subplots(nrows =2, ncols = 1, figsize = (8,2))
axes[0].plot(y,x)
axes[1].plot(x,y)
plt.tight_layout()
```
